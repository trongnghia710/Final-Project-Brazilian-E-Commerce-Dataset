## **5. Order Reviews**

### **Extract**

```python
df_order_reviews = pd.read_csv("/content/drive/MyDrive/olist/olist_order_reviews_dataset.csv")
df_order_reviews.head()
```

```python
df_order_reviews.info()
```

### **Transform**

```python
df_order_reviews["review_creation_date"] = pd.to_datetime(df_order_reviews["review_creation_date"])
df_order_reviews["review_answer_timestamp"] = pd.to_datetime(df_order_reviews["review_answer_timestamp"])
```

```python
df_order_reviews["order_id"] = df_order_reviews["order_id"].astype("category")
df_order_reviews["review_score"] = df_order_reviews["review_score"].astype("int8")
```

```python
df_order_reviews = df_order_reviews.sort_values("review_answer_timestamp")

df_order_reviews = df_order_reviews.drop_duplicates(
    subset=["order_id"],
    keep="last"
)

print("Duplicates left:", df_order_reviews["order_id"].duplicated().sum())

```

## **6. Orders**

### **Extract**

```python
df_orders = pd.read_csv("/content/drive/MyDrive/olist/olist_orders_dataset.csv")
df_orders.head()
```

```python
df_orders.info()
```

### **Transform**

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    df_orders[col] = pd.to_datetime(df_orders[col])
```

```python
cat_cols = ["order_id", "customer_id", "order_status"]

for col in cat_cols:
    df_orders[col] = df_orders[col].astype("category")
```

```python
df_orders["order_id"].duplicated().sum()
```

## **7. Products**

### **Extract**

```python
df_products = pd.read_csv("/content/drive/MyDrive/olist/olist_products_dataset.csv")
df_products.head()
```

```python
df_products.info()
```

### **Transform**

```python
df_products["product_id"] = df_products["product_id"].astype("category")
df_products["product_category_name"] = df_products["product_category_name"].astype("category")
```

```python
meta_cols = ["product_name_lenght", "product_description_lenght", "product_photos_qty"]
df_products[meta_cols] = df_products[meta_cols].fillna(0)
```

```python
size_cols = ["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]

for col in size_cols:
    df_products[col] = df_products[col].fillna(df_products[col].median())
```

```python
df_products["product_id"].duplicated().sum()
```
